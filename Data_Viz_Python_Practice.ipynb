# Basics

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")
plt.xlabel("x")
plt.ylabel("y")
plt.title("x vs y")

plt.subplot(1,2,2)
plt.plot(y,x,"g")
plt.xlabel("y")
plt.ylabel("x")
plt.title("y vs x")
plt.show()
```

# Object oriented methods

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("X vs Y")
```

```python
fig1 = plt.figure()
axes1 = fig1.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig1.add_axes([0.2, 0.4, 0.3, 0.4])


axes1.plot(x,y)
axes2.plot(y,x, "r")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
x = [10, 100, 1000,5000]
y = [10, 100, 1000, 10000]
```

```python
#Function method
plt.plot(x,y)
plt.title("1")
plt.xlabel("x")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.subplot(1,2,2)
plt.show()
```

```python
#Object oriented method 01
fig = plt.figure()
axe1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axe1.plot(x,y)
axe1.set_title("Title")
axe1.set_xlabel("x")
plt.show()
```

```python
#Object oriented method 02
fig,axes = plt.subplots(2,1, figsize=(8, 5), dpi =100)
axes[0].plot(x,y)
plt.show()
plt.tight_layout()
#fig.savefig("my_picture.png", dpi = 200)
```
